# clasificador
## para el clasificador se ingresara un nuevo arreglo aleatoriamente generado y posteriormente se calcularan sus medias y esperanzas por parejas lo que nos dara una ralacion con los datos de los conjuntos previamente analizados para su posterior clasificacion.

```python
seed(4)
valuesnew = randint(0, 100, 50)
median = valuesnew.mean()
varn = valuesnew.var()
stdn = valuesnew.std()
```

```python
bn = np.arange(2)
#bn = np.ones((1,2))
a= np.arange(50)
for i in range(49):
 #  bn[(0,0)]=valuesnew[0,i]
    bn[0]=valuesnew[i]
   #bn[0,0]=valuesnew[0,i]
    bn[1]=valuesnew[i+1]
    m = bn.mean()
    v = bn.var()
    d = bn.std()
    if m<= media or v<=var or d<=std:
        a[i]= 1
    else:
        a[i]=0
    if i ==49 and m<=media:
        a[49]=1
    else:
        a[49]=0
print(a)
```

```python
hist, bin_edges = np.histogram(a)
n, bins, patches = plt.hist(x=a, bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=0.85)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Value')
plt.ylabel('Frequency')
#plt.text(23, 45, r'$\mu=15, b=3$')
maxfreq = n.max()
# Set a clean upper y-axis limit.
plt.ylim(top=np.ceil(maxfreq / 10) * 10 if maxfreq % 10 else maxfreq + 10)
plt.show()
```
